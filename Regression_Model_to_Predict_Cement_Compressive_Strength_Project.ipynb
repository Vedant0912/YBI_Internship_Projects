```python
# import library
import pandas as pd
import seaborn
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns 
```

```python
# import data
cement = pd.read_csv('https://github.com/ybifoundation/Dataset/raw/main/Concrete%20Compressive%20Strength.csv')
```

```python
# view data
cement.head()
```

```python
# info of data
cement.info()
```

```python
# summary statistics
cement.describe()
```

```python
# check for missing value
cement.isnull().sum()
```

```python
# check for categories
cement.nunique()
```

```python
# visualize pairplot
sns.pairplot(cement)
```

```python
# columns name
cement.columns
```

```python
# define y
y = cement['Concrete Compressive Strength(MPa, megapascals) ']
```

```python
# define X
X = cement[['Cement (kg in a m^3 mixture)',
       'Blast Furnace Slag (kg in a m^3 mixture)',
       'Fly Ash (kg in a m^3 mixture)', 'Water (kg in a m^3 mixture)',
       'Superplasticizer (kg in a m^3 mixture)',
       'Coarse Aggregate (kg in a m^3 mixture)',
       'Fine Aggregate (kg in a m^3 mixture)', 'Age (day)']]
     
```

```python
# split data
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y, train_size=0.75, random_state=2529)
```

```python
# verify shape
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
# select model
from sklearn.linear_model import LinearRegression
model=LinearRegression()
```

```python
# train model
model.fit(X_train,y_train)
```

```python
# predict with model
y_pred=model.predict(X_test)
```

```python
# model evaluation
from sklearn.metrics import mean_absolute_error,mean_squared_error,mean_absolute_percentage_error
```

```python
# model MAE
mean_absolute_error(y_test,y_pred)
```

```python
# model MAPE
mean_absolute_percentage_error(y_test,y_pred)
```

```python
# model MSE
mean_squared_error(y_test,y_pred)
```
